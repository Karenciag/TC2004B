**Obtencion de Estadisticas Descriptivas**
El siguente código es extraido de los datos de "Estadísticas de los jugadores de la liga MX" de la página https://www.msn.com/es-mx/deportes/futbol/liga-mx/estadisticas-jugadores


```
import pandas as pd

# Obten los datos de la tabla HTML
url = 'https://www.msn.com/es-mx/deportes/futbol/liga-mx/estadisticas-jugadores'
html = pd.read_html(url, header = 0)
df = html[0]

raw = df

# Data cleaning
raw = df.set_axis(['#','Nombre','Equipo','Ab','PJ','PI','Min','Gol','Asis','Tiros','TP'], axis=1)
#raw = df2020.set_axis(['#','Nombre','Equipo','Ab','PJ','PI','Min','Gol','Asis','Tiros','TP'], axis='columns')
raw
```

### **GroupBy() function**

**¿Qué jugador de Tigres tuvo más minutos en cancha?**

```
Tigres = df.groupby('Equipo').get_group('Tigres UANL')
Tigres
```

**¿Cuáles son las estadísticas por partido jugado?**



```
df.groupby('PJ').PJ.describe()
```

**Muestra los equipos de Nuevo León.**

```
NL = ['Tigres UANL', 'Monterrey']
Pos = df[df['Equipo'].isin(NL)]
Pos
```
